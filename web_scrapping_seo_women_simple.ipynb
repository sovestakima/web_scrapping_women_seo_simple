```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import datetime as dt
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_women_CEOs_of_Fortune_500_companies'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
table_titles = soup.find_all('th')
table_titles
```

```python
table_titles = [title.text.strip() for title in table_titles]
```

```python
df = pd.DataFrame(columns = table_titles)
df
```

```python
table_data = soup.find_all('tr')[1:]
table_data
```

```python
for column in table_data:
    row_data = column.find_all('td')
    row_data = [row.text.strip() for row in row_data]
    
    length = len(df)
    df.loc[length] = row_data
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.rename(columns = lambda x: x.replace(' ','_').lower())
```

```python
df.head()
```

```python
df['fortune_500_rank'] = df['fortune_500_rank'].astype('int')
df['start_date'] = pd.to_datetime(df['start_date'])
```
